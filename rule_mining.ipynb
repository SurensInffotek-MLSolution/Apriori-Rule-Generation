```python
#convert the heading as the first row in transactions
#we can do that by reading the file by mentioning specifically
store_data = pd.read_csv('C:\\Users\\USER\\Desktop\\sinfotech\\Association Rule Mining\\store_data.csv', header=None)
```

```python
store_data.head()
```

```python
#converting the pandas data frame to a list of lists, since the apriori algorithm works in only that format as input
records = []
for i in range(0 ,len(store_data)):
    records.append([str(store_data.values[i,j]) for j in range(0,20)])
```

```python
#to check the resultant list of lists
records[1]
```

```python
rules = apriori(records, min_support = 0.0045, min_confidence = 0.2, min_lift = 2)
rules_results = list(rules)
#here support was taken as 0.0045 because we want minimum 35 sales per week and 5 sales per day of one particular item
```

```python
rules_results[0]
#to view the rules in a easy to understand format
#we can do this

for item in rules_results:
    pair = item[0]
    new_pair = [x for x in pair]
    print("Rule: " + new_pair[0] + " -> " + new_pair[1])
    
    print("Support: " + str(item[1]))
    print("confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("==============================")
```
